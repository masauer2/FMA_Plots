```python
from FMAPlots.colors import Colors
from FMAPlots.fes import Fes
```

```python
workingDir = "/scratch/masauer2/FMA_KEMP/KE_ML/metad_study/template_R1_metad/06-converged"
subWorkingDir = "07-reweight"
fName = "REWEIGHT-distances"

cv_x, cv_y, weights = Fes.import_data(workingDir, subWorkingDir, fName, 20)

xmin, xmax, ymin, ymax = 1.15, 1.35, 1.7, 1.95
histograms = Fes.get_histograms(cv_x, cv_y, weights, xmin, xmax, ymin, ymax)

probability, free_energy = Fes.get_probability(histograms, 20)
```
